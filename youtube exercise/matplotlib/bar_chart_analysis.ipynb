```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('program_lanauge_data.csv')
data.head()
```

```python
lan_count = Counter()
```

```python
langauges = data['LanguagesWorkedWith']

for row in langauges:
    lan_count.update(row.split(';'))
```

```python
lan_count_d = dict(lan_count)
```

```python
# lan = lan_count_d.keys()
# count =lan_count_d.values()
```

```python
df = pd.DataFrame(lan_count_d.items(), columns=['Pro_lan', 'Popularity'])
df = df.sort_values('Popularity', ascending = False)
lan = df['Pro_lan']
count =df['Popularity']
```

```python
plt.figure(figsize=(16,9))
plt.bar(lan, count)
plt.xlabel('Programming Language', fontsize= 15)
plt.xticks(rotation = 90)
plt.title('Programming Language Popularity',fontsize =25);
```

```python
lan2 = lan.to_list()
count2 = count.to_list()
lan2.reverse()
count2.reverse()
```

```python
plt.figure(figsize=(16,9))
plt.xkcd() ##comics style
plt.barh(lan2, count2, color = 'orange')
plt.ylabel('Programming Language', fontsize= 15)
plt.xticks(rotation = 90)
plt.title('Programming Language Popularity',fontsize =25)

plt.tight_layout();
```
